```python
#
# Example file for working with timedelta objects
# LinkedIn Learning Python course by Joe Marini
#
from datetime import datetime, timedelta, date
```

```python
# construct a basic timedelta and print it

print(timedelta(days=365, hours=5, minutes=1))
```

```python
# print today's date
now = datetime.now()
print(f'Today is: {now}')

# print today's date one year from now
print(f'One year from now is going to be: {now+timedelta(days=365)}')
```

```python

# create a timedelta that uses more than one argument
print(f'In 3 weeks and 4 days and 8 hours, it will be: {now+timedelta(\
  weeks=3, days=4, hours=8)}')
```

```python

# calculate the date 1 week ago, formatted as a string
# print(f'One week ago was: {now+timedelta(weeks=-1 )}')
t = now+timedelta(weeks=-1)
s = t.strftime('%A %B %d, %Y')
print(f'One week ago was: {s}')
```

```python
### How many days until April Fools' Day?
# t = date.today()
t = date(2026, 5, 1)
afd = date(t.year, 4, 1)

if t > afd:
  print(f'April fools day was {(t-afd).days} days ago')
  afdn = date(t.year+1, 4, 1)
  print(f"April fool's day of the next year is goint to be in {(afdn-t).days} days")
else:
  print(f'April fools day is going to be in {(afd-t).days} days')
```
