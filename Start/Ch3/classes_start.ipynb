```python
class Vehicle:
    def __init__(self, bodystyle):
        self.bodystyle = bodystyle

    def drive(self, speed):
        self.mode = "driving"
        self.speed = speed

class Car(Vehicle):
    def __init__(self, enginetype):
        super().__init__("Car")
        self.enginetype = enginetype
        self.wheels = 4
        self.doors = 4
    
    def drive(self, speed):
        super().drive(speed)
        print(f"Driving my {self.enginetype} car at {self.speed} mph")

class Motorcyle(Vehicle):
    def __init__(self, enginetype, hassidecar):
        super().__init__("Motorcycle")
        self.enginetype = enginetype
        if hassidecar:
            self.wheels = 3
        else:
            self.wheels = 2
        self.doors = 0
        self.speed = 0
    
    def drive(self, speed):
        super().drive(speed)
        print(f"Driving my {self.enginetype} motorcycle at {self.speed} mph")

car1 = Car("gas")
car2 = Car("electric")
mv1 = Motorcyle("gas", True)
```

```python
print(car1.wheels)
print(car2.enginetype)
print(mv1.wheels)
print(car2.doors)
```

```python
car1.drive(50)
car2.drive(25)
mv1.drive(75) 
```
