```python

# function that takes parameters
def hello_func(greeting):
    print('Hey')
    name = input("What is your name? ")
    print(greeting, name)

hello_func("Welcome to Python programming,")
```

```python

# function that returns a value

def cube_func(x):
    return x * x * x
print(cube_func(3))
```

```python

# function with default value for an parameter
def hello_func(greeting, name=None):
    print('Hey')
    if name == None:
        name = input("What is your name? ")
    print(greeting, name)

hello_func("Welcome to Python programming,", 'Joe')
```

```python

# function with variable number of parameters
def multi_add(start, *args):
    # result = start
    for x in args:
        start += x
    return start

print(multi_add(1, 2, 3, 4, 5, -4, 12))
```
